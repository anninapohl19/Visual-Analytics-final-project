### Data Cleaning - Games Table

```python
import pandas as pd
```

#### Import and basic information

```python
games = pd.read_csv('games.grivg.csv')
print(games.head())
print(games.info())
print(games.describe())
```

#### Drop the last column

```python
games = games.drop(['Unnamed: 27'], axis = 1)
```

#### Check for duplicates

```python
duplicates_char = games.duplicated()
duplicates_char.value_counts()
duplicates_char.sum()
```

```python
games.nunique()
```

#### Check for null values and handle them

```python
games.isnull().sum()
```

Not all games are part of a Series. Therefore the Null values are totally fine and nothing is changed. 

BUT: the missing values in the ratings are filled in with the average of the other available ratings.

##### Destructorid column

```python
games[games['Destructoid'].isnull()]
```

```python
games['Destructoid'] = games['Destructoid'].fillna(games['Avg_Reviews'])
```

##### IGN column

```python
games[games['IGN'].isnull()]
```

```python
games['IGN'] = games['IGN'].fillna(games['Avg_Reviews'])
```

##### GameSpot column

```python
games[games['GameSpot'].isnull()]
```

```python
games['GameSpot'] = games['GameSpot'].fillna(games['Avg_Reviews'])
```

```python
games.isnull().sum()
```

#### Trim leading and trailing space of all string columns

```python
games['Game_Id'] = games['Game_Id'].str.strip()
games['Title'] = games['Title'].str.strip()
games['Series'] = games['Series'].str.strip()
games['Genre'] = games['Genre'].str.strip()
games['Sub-genre'] = games['Sub-genre'].str.strip()
games['Developer'] = games['Developer'].str.strip()
games['Publisher'] = games['Publisher'].str.strip()
games['Country'] = games['Country'].str.strip()
games['Platform'] = games['Platform'].str.strip()
games['Customizable_main'] = games['Customizable_main'].str.strip()
games['Criteria'] = games['Criteria'].str.strip()
games['Director'] = games['Director'].str.strip()
print(games)
```

#### Converting types

```python
# check the types
games.dtypes
```

##### Release column into datetime

```python
games['Release'] = games['Release'].str.replace('Jan', '01')
games['Release'] = games['Release'].str.replace('Feb', '02')
games['Release'] = games['Release'].str.replace('Mar', '03')
games['Release'] = games['Release'].str.replace('Apr', '04')
games['Release'] = games['Release'].str.replace('May', '05')
games['Release'] = games['Release'].str.replace('Jun', '06')
games['Release'] = games['Release'].str.replace('Jul', '07')
games['Release'] = games['Release'].str.replace('Aug', '08')
games['Release'] = games['Release'].str.replace('Sep', '09')
games['Release'] = games['Release'].str.replace('Oct', '10')
games['Release'] = games['Release'].str.replace('Nov', '11')
games['Release'] = games['Release'].str.replace('Dec', '12')
games
```

```python
games['Release'] = games['Release'].astype(str)
```

```python
new = games['Release'].str[:3] + '20' + games['Release'].str[3:]
games['Release'] = new
games['Release'] = games['Release'].astype('datetime64[ns]')
games
```

##### Percentage_non_male column into float

```python
games['Percentage_non_male'] = games['Percentage_non_male'].astype(str)
for i in range(len(games['Percentage_non_male'])):
    if len(games['Percentage_non_male'][i]) == 3:
        games['Percentage_non_male'][i] = '0.' + games['Percentage_non_male'][i].rstrip('%')
    else:
        games['Percentage_non_male'][i] = '0.0' + games['Percentage_non_male'][i].rstrip('%')
games['Percentage_non_male'] = games['Percentage_non_male'].astype('float64')
games['Percentage_non_male']
```

##### Team_percentage column into float

```python
games['Team_percentage'] = games['Team_percentage'].astype(str)
for i in range(len(games['Team_percentage'])):
    if len(games['Team_percentage'][i]) == 3:
        games['Team_percentage'][i] = '0.' + games['Team_percentage'][i].rstrip('%')
    else:
        games['Team_percentage'][i] = '0.0' + games['Team_percentage'][i].rstrip('%')
games['Team_percentage'] = games['Team_percentage'].astype('float64')
games['Team_percentage']
```
